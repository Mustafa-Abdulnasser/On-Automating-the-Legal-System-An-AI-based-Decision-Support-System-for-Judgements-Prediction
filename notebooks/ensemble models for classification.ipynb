# Loading Test Data

```python
import ast
file_path = "/GP/tobert.txt"
tobert = []
with open(file_path, 'r') as file:
    for line in file:
        line = line.strip()
        float_value = ast.literal_eval(line)[0]
        tobert.append(float_value)

file_path = "/GP/tail.txt"
tail = []
with open(file_path, 'r') as file:
    for line in file:
        line = line.strip()
        float_value = ast.literal_eval(line)[0]
        tail.append(float_value)

file_path = "/GP/head.txt"
head = []
with open(file_path, 'r') as file:
    for line in file:
        line = line.strip()
        float_value = ast.literal_eval(line)[0]
        head.append(float_value)

file_path = "/GP/max.txt"
maxAlg = []
with open(file_path, 'r') as file:
    for line in file:
        line = line.strip()
        float_value = ast.literal_eval(line)[0]
        maxAlg.append(float_value)

file_path = "/GP/mean.txt"
mean = []
with open(file_path, 'r') as file:
    for line in file:
        line = line.strip()
        float_value = ast.literal_eval(line)[0]
        mean.append(float_value)

file_path = "/GP/robert.txt"
robert = []
with open(file_path, 'r') as file:
    for line in file:
        line = line.strip()
        float_value = ast.literal_eval(line)[0]
        robert.append(float_value)
        
file_path = "/GP/target.txt"
target = []
with open(file_path, 'r') as file:
    for line in file:
        line = line.strip()
        float_value = ast.literal_eval(line)[0]
        target.append(float_value)
        

# Combine all lists into a single list
lists = [tobert,tail,maxAlg,mean,robert,head]
```

```python
# Get the length of the lists
lengthOfLists = len(lists[0]) 
noOfLists = len(lists) 

# Generate all combinations for the same index
combinations = []
keys={0:"tobert",1:"tail",2:"max",3:"mean",4:"robert",5:"head"}
results={}

def getCombinOfSize2():
    for no1 in range (0,noOfLists):
        for no2 in range(no1+1 , noOfLists):
            for i in range(0,lengthOfLists):
                if(((lists[no1][i]+lists[no2][i])/2.0)>=0.5):
                    combinations.append(1)
                else:
                    combinations.append(0)
            results[keys[no1]," & ",keys[no2]]=combinations.copy()
            combinations.clear()
    
                
def getCombinOfSize3():
    for no1 in range (0,noOfLists):
        for no2 in range(no1+1 , noOfLists):
            for no3 in range(no2+1,noOfLists):
                for i in range(0,lengthOfLists):
                    if(((lists[no1][i]+lists[no2][i]+lists[no3][i])/3.0)>=0.5):
                        combinations.append(1)
                    else:
                        combinations.append(0)
                results[keys[no1]," & ",keys[no2]," & ",keys[no3]]=combinations.copy()
                combinations.clear()
                    
def getCombinOfSize4():
    for no1 in range (0,noOfLists):
        for no2 in range(no1+1 , noOfLists):
            for no3 in range(no2+1,noOfLists):
                for no4 in range(no3+1,noOfLists):
                    for i in range(0,lengthOfLists):
                        if(((lists[no1][i]+
                             lists[no2][i]+
                             lists[no3][i]+
                            lists[no4][i])/4.0)>=0.5):
                            combinations.append(1)
                        else:
                            combinations.append(0)
                    results[keys[no1]," & ",keys[no2]," & ",keys[no3]," & ",keys[no4]]=combinations.copy()
                    combinations.clear()

def getCombinOfSize5():
    for no1 in range (0,noOfLists):
        for no2 in range(no1+1 , noOfLists):
            for no3 in range(no2+1,noOfLists):
                for no4 in range(no3+1,noOfLists):
                    for no5 in range(no4+1,noOfLists):
                        for i in range(0,lengthOfLists):
                            if(((lists[no1][i]+
                                 lists[no2][i]+
                                 lists[no3][i]+
                                lists[no4][i]+
                                lists[no5][i])/5.0)>=0.5):
                                combinations.append(1)
                            else:
                                combinations.append(0)
                        results[keys[no1]," & ",
                                keys[no2]," & ",
                                keys[no3]," & ",
                               keys[no4]," & ",
                               keys[no5]]=combinations.copy()
                        combinations.clear()

def getCombinOfSize6():
    for no1 in range (0,noOfLists):
        for no2 in range(no1+1 , noOfLists):
            for no3 in range(no2+1,noOfLists):
                for no4 in range(no3+1,noOfLists):
                    for no5 in range(no4+1,noOfLists):
                        for no6 in range(no5+1,noOfLists):
                            for i in range(0,lengthOfLists):
                                if(((lists[no1][i]+
                                     lists[no2][i]+
                                     lists[no3][i]+
                                    lists[no4][i]+
                                    lists[no5][i]+
                                    lists[no6][i])/6.0)>=0.5):
                                    combinations.append(1)
                                else:
                                    combinations.append(0)
                            results[keys[no1]," & ",
                                keys[no2]," & ",
                                keys[no3]," & ",
                               keys[no4]," & ",
                               keys[no5]," & ",
                                   keys[no6]]=combinations.copy()
                            combinations.clear()

getCombinOfSize2()
getCombinOfSize3()
getCombinOfSize4()
getCombinOfSize5()
getCombinOfSize6()
```

```python
from sklearn.metrics import f1_score, recall_score, precision_score
for q in range(0,noOfLists):
    print(keys[q])
    lists[q]= [1 if (i >= 0.5) else 0 for i in lists[q]]
    f1 = f1_score(target, lists[q], average='macro')
    print(f1)
for key in results.keys():
    print(key)
    f1 = f1_score(target, results[key], average='macro')
    print(f1)
```

```python
# Get the length of the lists
lengthOfLists = len(lists[0]) 
noOfLists = len(lists) 

# Generate all combinations for the same index
combinations = []
keys={0:"tobert",1:"tail",2:"max",3:"mean",4:"robert",5:"head"}
results={}

def getCombinOfSize2():
    for no1 in range (0,noOfLists):
        for no2 in range(no1+1 , noOfLists):
            for i in range(0,lengthOfLists):
                if(lists[no1][i]>=0.5 and lists[no2][i]>=0.5):
                    combinations.append(1)
                elif(lists[no1][i] < 0.5 and lists[no2][i] < 0.5):
                    combinations.append(0)
                elif(lists[no1][i] >= 0.5): 
                    if(1 - lists[no1][i] > lists[no2][i] - 0):
                        combinations.append(0)
                    else:
                        combinations.append(1)
                elif(lists[no2][i] >= 0.5):
                    if(1 - lists[no2][i] > lists[no1][i] - 0):
                        combinations.append(0)
                    else:
                        combinations.append(1)            
            results[keys[no1]," & ",keys[no2]]=combinations.copy()
            combinations.clear()
    
                
def getCombinOfSize3():
    for no1 in range (0,noOfLists):
        for no2 in range(no1+1 , noOfLists):
            for no3 in range(no2+1,noOfLists):
                for i in range(0,lengthOfLists):
                    counter=round(lists[no1][i])+round(lists[no2][i])+round(lists[no3][i])
                    if(counter>=2):
                        combinations.append(1)
                    else:
                        combinations.append(0)

                results[keys[no1]," & ",keys[no2]," & ",keys[no3]]=combinations.copy()
                combinations.clear()
                    
def getCombinOfSize4():
    for no1 in range (0,noOfLists):
        for no2 in range(no1+1 , noOfLists):
            for no3 in range(no2+1,noOfLists):
                for no4 in range(no3+1,noOfLists):
                    for i in range(0,lengthOfLists):
                        counter=round(lists[no1][i])+round(lists[no2][i])+round(lists[no3][i])+round(lists[no4][i])
                        if(counter>2):
                            combinations.append(1)
                        elif(counter<2):
                            combinations.append(0)
                        else:
                            if(((lists[no1][i]+lists[no2][i]+lists[no3][i]+lists[no4][i])/4)>=0.5):
                                   combinations.append(1)
                            else:
                                   combinations.append(0)

                    results[keys[no1]," & ",keys[no2]," & ",keys[no3]," & ",keys[no4]]=combinations.copy()
                    combinations.clear()

def getCombinOfSize5():
    for no1 in range (0,noOfLists):
        for no2 in range(no1+1 , noOfLists):
            for no3 in range(no2+1,noOfLists):
                for no4 in range(no3+1,noOfLists):
                    for no5 in range(no4+1,noOfLists):
                        for i in range(0,lengthOfLists):
                            counter=round(lists[no1][i])+round(lists[no2][i])+round(lists[no3][i])+round(lists[no4][i])+round(lists[no5][i])
                            if(counter>=3):
                                combinations.append(1)
                            else:
                                combinations.append(0)

                        results[keys[no1]," & ",
                                keys[no2]," & ",
                                keys[no3]," & ",
                               keys[no4]," & ",
                               keys[no5]]=combinations.copy()
                        combinations.clear()

def getCombinOfSize6():
    for no1 in range (0,noOfLists):
        for no2 in range(no1+1 , noOfLists):
            for no3 in range(no2+1,noOfLists):
                for no4 in range(no3+1,noOfLists):
                    for no5 in range(no4+1,noOfLists):
                        for no6 in range(no5+1,noOfLists):
                            for i in range(0,lengthOfLists):
                                counter=round(lists[no1][i])+round(lists[no2][i])+round(lists[no3][i])
                                +round(lists[no4][i])+round(lists[no5][i])+round(lists[no6][i])
                                if(counter>3):
                                    combinations.append(1)
                                elif(counter<3):
                                    combinations.append(0)
                                else:
                                    if(((lists[no1][i]+
                                         lists[no2][i]+
                                         lists[no3][i]+
                                        lists[no4][i]+
                                        lists[no5][i]+
                                        lists[no6][i])/6.0)>=0.5):
                                        combinations.append(1)
                                    else:
                                        combinations.append(0)
                            results[keys[no1]," & ",keys[no2]," & ",keys[no3]," & ",keys[no4]," & ",keys[no5]," & ",keys[no6]]=combinations.copy()
                            combinations.clear()

getCombinOfSize2()
getCombinOfSize3()
getCombinOfSize4()
getCombinOfSize5()
getCombinOfSize6()
```
