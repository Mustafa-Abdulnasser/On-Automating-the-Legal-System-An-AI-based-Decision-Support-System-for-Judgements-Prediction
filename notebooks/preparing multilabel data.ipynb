```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('data/data.csv')
df
```

```python
df_vioa = df[['Text','VIOA']]
df_vioa.head()
```

```python
df_vioa.info()
```

```python
df_vioa.VIOA.value_counts()
```

```python
VIOA = set()
vioa = list()
for item in df_vioa.VIOA:
    item = item.replace('[','').replace(']','').replace('\'','').replace(' ','').split(',')
    vioa.append(item)
    for x in item:
        VIOA.add(x)
VIOA.remove('')
print(VIOA)
print(len(VIOA))
```

```python
df_vioa['VIOA'] = vioa
df_vioa.head()
```

```python
df_vioa.info()
```

```python
df_vioa.VIOA.value_counts()
```

```python
df2 = pd.DataFrame(np.nan , index = range(0,len(vioa)), columns = VIOA)
```

```python
df2
```

```python
for row in range(0,len(vioa)):
    for item in vioa[row]:
        if item != '':
            df2[item][row] = 1
df2
```

```python
df2.iloc[3]
```

```python
df2.fillna(0,inplace = True)
df2
```

```python
articles = pd.concat([df['Text'],df2],axis = 1)
articles
```
